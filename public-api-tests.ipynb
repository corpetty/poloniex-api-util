```python
%matplotlib inline

import matplotlib as plt
import requests
import json
import pandas as pd
```

```python
# We want to keep the channel alive as we make requests to poloniex
s = requests.session()
```

All API calls can be found in documentation at https://poloniex.com/support/api/

```python
url = 'https://poloniex.com/public?command=returnTicker&currencyPair=BTC_ETH'
req = s.get(url)
df = pd.read_json(req.text)
```

## Let's look at the order book

```python
eth_book_url = 'https://poloniex.com/public?command=returnOrderBook&currencyPair=BTC_ETH&depth=500'
ethbook = pd.read_json(s.get(eth_book_url).text)
```
